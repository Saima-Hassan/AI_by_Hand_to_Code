```python
import numpy as np

# Input values
x = np.array([2, 1, 3])  # Input vector
x = x.reshape(-1,1)
# Weight matrix for 4 neurons (each row represents the weights for a neuron)
w1 = np.array([[1, -1, 1], 
              [1, 1, 0], 
              [0, 1, 1], 
              [1, 0, 1]])
# Bias vector for 4 neurons
b1 = np.array([-5, 0, 1, -2])  # Bias vector
b1= b1.reshape(-1,1)
```

```python
# ReLU activation function
def relu(x):
    return np.maximum(0,x)
```

```python
   # Linear combination (dot product + bias)
z1_pre = np.dot(w1,x) + b1
print(f"pre-activation output: \n {z1_pre}")

# Apply ReLU activation function
h = relu(z1)

# Print the output for all neurons
print(f"Output after applying RELU: \n {h}")
```

```python
w2 = np.array([[1, 1, -1, 0], 
               [0, 0, 1, -1]])

# Bias for the output layer neurons
b2 = np.array([[0], [1]]) 

z2_pre = np.dot( w2, z1) + b2
print(f"pre-activation output: \n {z2_pre}")

# Apply ReLU activation function
z2 = relu(z2_pre)

# Print the output for all neurons
print(f"Output after applying RELU: \n {z2}")
```
