```python
import numpy as np

# Input values
X = np.array([[2, 1, 0],
             [1, 1, 1],
             [3, 0, 1]])  # Input matrix

# Weight matrix for 4 neurons (each row represents the weights for a neuron)
w1 = np.array([[1, -1, 1], 
              [1, 1, 0], 
              [0, 1, 1], 
              [1, 0, 1]])
# Bias vector for 4 neurons
b1 = np.array([[-5], [0], [1], [-2]])  # Bias vector

# ReLU activation function
def relu(x):
    return np.maximum(0,x)
```

```python
def hidden_layer_neuron(x, w, b):
    """
    x: input vector (input to the hidden layer)
    w: weights for the hidden layer (each row represents a neuron's weights)
    b: biases for the hidden layer (one bias per neuron)
    
    returns: output from the hidden layer neurons after ReLU activation
    """
    z = np.dot(w, x) + b

    # Apply ReLU activation function
    output = relu(z)
    # Return the list of outputs for all neurons
    
    return (output)
```

```python
z1 = hidden_layer_neuron(X, w1, b1)
# Print the output for all neurons
print(f"Output of the hidden layer neurons: \n{z1}")
```

```python
# Weights for the output layer (2 neurons in output layer)
w2 = np.array([[1, 1, -1, 0], 
               [0, 0, 1, -1]])

# Bias for the output layer neurons
b2 = np.array([[0], [1]])  # Bias for 2 output neurons

#input to the 2nd layer is "z1"

# 2. Calculate the output layer output
# Output layer calculation: (z1, w2) + b2
z2=hidden_layer_neuron(z1, w2, b2)

# Print the output for outer neurons
print(f"Output of the outer layer neurons: \n{z2}")
```
